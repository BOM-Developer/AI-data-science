Overview
This project involves the development of a machine learning model to predict heart disease using the RandomForestClassifier in Scikit-Learn. Heart_Disease_Prediction.ipynb: Jupyter notebook containing the code for data preprocessing, model training, evaluation, and serialization.

```python
import numpy as np
import pandas as pd

# Assuming the heart disease dataset has similar features to the common UCI Heart Disease dataset
# These features might include age, sex, chest pain type (cp), resting blood pressure (trestbps),
# cholesterol (chol), fasting blood sugar (fbs), resting electrocardiographic results (restecg),
# maximum heart rate achieved (thalach), exercise induced angina (exang),
# oldpeak = ST depression induced by exercise relative to rest, the slope of the peak
# exercise ST segment (slope), number of major vessels (0-3) colored by fluoroscopy (ca),
# and thal: 3 = normal; 6 = fixed defect; 7 = reversible defect.

# Generate random data for these features
np.random.seed(42)  # for reproducibility
sample_data = {
    'age': np.random.randint(29, 77, 10),  # age between 29 and 76
    'sex': np.random.randint(0, 2, 10),  # sex either 0 or 1
    'cp': np.random.randint(0, 4, 10),  # chest pain type between 0 and 3
    'trestbps': np.random.randint(94, 200, 10),  # resting blood pressure
    'chol': np.random.randint(126, 564, 10),  # serum cholesterol
    'fbs': np.random.randint(0, 2, 10),  # fasting blood sugar > 120 mg/dl
    'restecg': np.random.randint(0, 2, 10),  # resting electrocardiographic results
    'thalach': np.random.randint(71, 202, 10),  # maximum heart rate achieved
    'exang': np.random.randint(0, 2, 10),  # exercise induced angina
    'oldpeak': np.random.uniform(0.0, 6.2, 10),  # ST depression induced by exercise relative to rest
    'slope': np.random.randint(0, 3, 10),  # the slope of the peak exercise ST segment
    'ca': np.random.randint(0, 4, 10),  # number of major vessels colored by fluoroscopy
    'thal': np.random.randint(1, 4, 10)  # 3 = normal; 6 = fixed defect; 7 = reversible defect
}

# Create a DataFrame
random_df = pd.DataFrame(sample_data)

random_df
```
