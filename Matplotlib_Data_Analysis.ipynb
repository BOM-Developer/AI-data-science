# **Real Life Project Ploting**

### **pyplot with Object Oriented Method**

```python
heart = pd.read_csv("/content/heart.csv")
heart
```

```python
heart.head(4)
```

```python
over_50 = heart[heart["age"]>50]
over_50
```

```python
# Let's apply pyplot methods
over_50 .plot(kind='scatter',
              x='age',
              y='chol',
              c='target');
```

### **OOM(object oriented methods**

```python
fig, ax = plt.subplots(figsize=(8, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',  # Specify the y column
             c='target',
             ax=ax);
```

### **Life Cycle of OO method**
* Creating plot area


```python
fig, ax = plt.subplots(figsize=(6,4))
# to plot the data let create a scatter plot
scatter_plot =ax.scatter(x=heart['age'],
                        y=heart['chol'],
                         c=heart['target']);

# le customiza the data for better understanding

ax.set(title='Heart disease problem',
       xlabel='Age',
       ylabel='Cholistrole')

# Legend setting
ax.legend(*scatter_plot.legend_elements(), title='target')
scatter_plot;
```

### **Creating two plots on a sinhle figure**

```python
import matplotlib.pyplot as plt

# Assuming 'over_50' is your DataFrame containing the data

fig, (plot1, plot2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 8),sharex=True)

# Plotting the first scatter plot
scatter_plot1 = plot1.scatter(x=over_50['age'],
                              y=over_50['chol'],
                              c=over_50['target'])

# Customizing plot 1
plot1.set(title='Heart Disease Problem with Cholestrol',
          # xlabel='Age',
          ylabel='Cholesterol')

# Legend setting for plot 1
plot1.legend(*scatter_plot1.legend_elements(), title='Survivalchances=1,Lesschances=0')

# Plotting the second scatter plot
scatter_plot2 = plot2.scatter(x=over_50['age'],
                              y=over_50['trestbps'],
                              c=over_50['target'])

# Customizing plot 2
plot2.set(title='Heart Disease with Resting BP',
          xlabel='Age',
          ylabel='Resting BP')

# Legend setting for plot 2
plot2.legend(*scatter_plot2.legend_elements(), title='Survivalchances=1, Lesschances=0')

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

### **Getting Mean Line**

```python
import matplotlib.pyplot as plt

# Assuming 'over_50' is your DataFrame containing the data

fig, (plot1, plot2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 8),sharex=True)

# Plotting the first scatter plot
scatter_plot1 = plot1.scatter(x=over_50['age'],
                              y=over_50['chol'],
                              c=over_50['target'])

# Customizing plot 1
plot1.set(title='Heart Disease Problem with Cholestrol',
          # xlabel='Age',
          ylabel='Cholesterol')

# Legend setting for plot 1
plot1.legend(*scatter_plot1.legend_elements(), title='Survivalchances=1,Lesschances=0')
plot1.axhline(y=over_50['chol'].mean(),linestyle="--")
# Plotting the second scatter plot
scatter_plot2 = plot2.scatter(x=over_50['age'],
                              y=over_50['trestbps'],
                              c=over_50['target'])


# Customizing plot 2
plot2.set(title='Heart Disease with Resting BP',
          xlabel='Age',
          ylabel='Resting BP')

# Legend setting for plot 2
plot2.legend(*scatter_plot2.legend_elements(), title='Survivalchances=1, Lesschances=0')
plot2.axhline(y=over_50['trestbps'].mean(),linestyle="--")
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
