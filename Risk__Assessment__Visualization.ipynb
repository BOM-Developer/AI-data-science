## **Visualizing Risk Landscapes**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the risk categories
risk_categories = {
    "Strategic": [
        "Technology/product", "Capacity", "Planning", "Industry",
        "Referrals/Revenue Growth", "Human Resources", "Public relations / publicity",
        "Disaster Recovery/Bus. Cont", "Intellectual Property", "Political",
        "Economic", "New business development", "Competitors", "Reimbursement"
    ],
    "Operational": [
        "Customer / Distributor satisfaction", "Product quality", "Information Technology",
        "Liquidity", "Production", "Obsolescence", "Sourcing (Resource Availability)",
        "Price", "Safety", "Cost"
    ],
    "Reporting and Compliance": [
        "Regulatory", "Financial reporting", "Tax", "Litigation", "Ethics and Integrity"
    ]
}

# Sample data for illustration (replace with your actual data)
prior_year_risks = {
    "Technology ": 3,
    "Capacity": 2,
    "Planning": 1,
    "Industry": 4,
    "Referrals": 3,
    "Human Resources": 2,
    "Public relations ": 1,
    "Disaster Recovery/Bus. Cont": 4,
    "Intellectual Property": 3,
    "Political": 2,
    "Economic": 1,
    "New business development": 4,
    "Competitors": 3,
    "Reimbursement": 2,
    "Customer satisfaction": 1,
    "Product quality": 4,
    "Information Technology": 3,
    "Liquidity": 2,
    "Production": 1,

    "Sourcing ": 3,

    "Safety": 1,
    "Cost": 4,
    "Regulatory": 3,
    "Financial reporting": 2,
    "Tax": 1,
    "Litigation": 4,
    "Ethics and Integrity": 3
}

# Convert data to DataFrame
df = pd.DataFrame(prior_year_risks.values(), index=prior_year_risks.keys(), columns=["Risk Score"])

# Pivot the data for heatmap plotting
df_pivot = df.pivot_table(columns=df.index)

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_pivot, cmap="YlGnBu", annot=True, fmt=".0f", linewidths=.5)
plt.title("Risk Heatmap for Prior Year")
plt.xlabel("Risk Categories")
plt.ylabel("Risk Factors")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
